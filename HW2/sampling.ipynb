```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean = np.zeros(2)
cov = np.array([[1, -0.75], [-0.75, 1]])
cov2 = np.array([[9, 0], [0, 1]])

s = np.random.multivariate_normal(mean, cov, size=2500)
print(s.shape)

rdm = np.random.normal(0, 1, size=(1000))
rdm2 = rdm * 4
cov_rdm = np.cov(rdm, rdm2)
print(cov_rdm)

plt.scatter(s[:,0], s[:,1], [1 for i in range(len(s))])
plt.xlim(-9, 9)
plt.ylim(-9, 9)
```
